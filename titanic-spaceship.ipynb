```python
spa = space[numerical]
spa[categ] = space[categ]

spa.dropna(inplace=True)

spa.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
label = LabelEncoder()
for i in col:
    spa[i] = label.fit_transform(spa[[i]])
```

```python
# does scaling effects the data or model train 
scaler = MinMaxScaler()
spa = spa.drop(['Name'], axis = 1)
for i in spa.columns:
    spa[i] = scaler.fit_transform(spa[[i]])
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
x = spa.drop('Transported', axis = 1)
y = spa['Transported']
 
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.3, random_state = 42)
```
